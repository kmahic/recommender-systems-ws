## Oppsummering av workshopen

| # | Lærdom |
|---|--------|
| 1 | **Globalt vs lokalt** — Bruk globale forklaringer for å forstå modellens overordnede atferd, lokale forklaringer for individuelle beslutninger |
| 2 | **Flere metoder** — Stol aldri på ett enkelt XAI-verktøy. Sammenlign SHAP, LIME, PDP, osv. |
| 3 | **Skepsis** — Forklaringer forklarer *modellen*, ikke *virkeligheten*. Sjekk for lekkasje, korrelasjoner, ustabilitet |
| 4 | **Subgrupper** — Globale mønstre kan skjule lokale reverseringer. Del alltid opp analysen per relevante grupper |
